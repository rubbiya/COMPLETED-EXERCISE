## k_nearest neighbour theory

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.head()
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:,-1].values
```

```python
x
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
x_train= sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
classifier.fit(x_train, y_train)
```

```python
print(classifier.predict(sc.transform([[30, 187000]])))
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
y_pred = y_pred.reshape(len(y_pred), 1)
y_test = y_test.reshape(len(y_test), 1)
```

```python
y_pred.shape
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print(cm)
print(acc)
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_train), y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:,0].min() - 10, stop = x_set[:, 0].max() + 10, step = 1),
                    np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 1))
plt.contourf(x1, x2, classifier.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),
            alpha = 0.75, cmap = ListedColormap(('red','green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(),x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set==j, 0], x_set[y_set ==j, 1], c= ListedColormap(('red','green'))(i),label = j)
    
plt.title('K-NN(Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## XGBoost

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Social_Network_Ads.csv')

dataset.head()

x = dataset.iloc[:, :-1].values
y = dataset.iloc[:,-1].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=0)

from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
x_train= sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)

print(cm)
print(acc)
```

```python
from sklearn.model_selection import cross_val_score
acc = cross_val_score(estimator = classifier, x = x_train, y=y_train, cv = 10)
print('Accuracy:{:0.2f}%'.format(acc.mean()*100))
print('Standard Deviation: {:0.2f}%'.format(acc.std()*100))
```
